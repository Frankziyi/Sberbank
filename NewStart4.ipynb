```python
import numpy as np
import pandas as pd
import xgboost as xgb
```

```python
train_df = pd.read_csv("./train.csv")
test_df = pd.read_csv("./test.csv")
```

```python
macro_df = pd.read_csv("./macro.csv")
```

```python
from numpy import nan
```

```python
train_df["kitch_sq"].ix[train_df["kitch_sq"] > 1000] = np.nan
test_df["kitch_sq"].ix[test_df["kitch_sq"] > 1000] = np.nan
```

```python
train_df["max_floor"].ix[train_df["max_floor"] > 98] = np.nan
test_df["max_floor"].ix[test_df["max_floor"] > 98] = np.nan
```

```python
train_df["state"].ix[train_df["state"] == 33] = np.nan
```

```python
dt = pd.to_datetime(macro_df['timestamp'])
```

```python
import datetime
help(datetime)
```

```python
dt = dt + datetime.timedelta(days=90)
```

```python
macro_df['timestamp'] = dt.dt.year.astype(str) + '/' + dt.dt.month.astype(str) + '/' + dt.dt.day.astype(str)
```

```python
macro_df['timestamp']
```

```python
train_df = train_df.replace({"no": 0, "yes": 1})
test_df = test_df.replace({"no": 0, "yes": 1})
```

```python
y_train = train_df['price_doc'].values
id_test = test_df['id']
```

```python
macro_df.drop(['child_on_acc_pre_school', 'modern_education_share', 'old_education_build_share'], axis=1, inplace=True)
```

```python
train_df2 = train_df.merge(macro_df, on='timestamp', how='left')
test_df2 = test_df.merge(macro_df, on='timestamp', how='left')
```

```python
train_df2.drop(['timestamp', 'id', 'sub_area', 'ecology', 'price_doc', 'product_type'], axis=1, inplace=True)
test_df2.drop(['timestamp', 'id', 'sub_area', 'ecology', 'product_type'], axis=1, inplace=True)
```

```python
xgb_params = {
    'eta': 0.05,
    'max_depth': 4,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'min_child_weight':1,
    'silent': 1,
    'seed':0
}
```

```python
dtrain = xgb.DMatrix(train_df2, y_train)
dtest = xgb.DMatrix(test_df2)
model = xgb.train(dict(xgb_params, silent=0), dtrain, num_boost_round= 480)
y_predict = model.predict(dtest)
output = pd.DataFrame({'id': id_test, 'price_doc': y_predict})
```
